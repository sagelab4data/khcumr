## 2.1 상관계수 (correlation)

* 상관계수 correlation는 두 변수간의 크기를 나타내주는 계수로서 -1 에서 1 사이의 값을 갖는다. 
* -1, 또는 1에 가까울수로 관계가 밀접하고, 0에 가까울 수록 서로 관계가 없음을 의미한다.
