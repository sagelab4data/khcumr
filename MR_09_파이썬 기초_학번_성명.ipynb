## **4.4 난수로 구성된 배열**


**[난수 random variable 생성]**

난수 random variable은 우연에 의해서만 값을 생성한다. (공정한 뽑기!)
단, 난수를 발생시키더라도 범위와 확률 조건을 부여 할 수 있다. 

넘파이의 서브모듈인 random 모듈에서는 다음과 같은 난수 생성기를 제공한다. 

---
* np.random 모듈에서 함수를 호출한다.

1. 0~1 사이에서 균등한 확률로 난수생성 : np.random.rand(난수 개수)
2. 시작과 끝 사이에서 균등한 확률로 난수 생성 : np.random.uniform (시작, 끝, 난수개수)
3. 시작과 끝 사이에서 균등한 확률로 정수 난수 생성 : np.random.randint (시작, 끝, 난수개수) 
4. 평균 0, 표준편차 1인 정규분포 확률에 따라서 난수 생성 : np.random.randn(난수개수)

5. 난수값을 고정 시키기 위해서 seed 값을 지정한다. : np.random.seed(시드값)
