# **3.데이터의 생성과 시각화**

코랩을 통해 파이썬의 풍부한 데이터 분석 라이브러리를 활용하여 데이터를 분석하고 시각화할 수 있습니다.

* 아래 코드 셀에서는 <strong>넘파이Numpy</strong>를 사용하여 임의의 데이터를 생성하고 <strong>매트플롯립matplotlib</strong>으로 이를 시각화합니다. 


* 직접 난수를 발생 시키고 이를 그래프로 표현해 봅니다. 


```
# 숫자 연산을 다루는 numpy와 그래프를 그리는 matplotlib을 읽어 들입니다.
import numpy as np
import matplotlib.pyplot as plt

# numpy를 이용하여 난수 random value를 발생 시킵니다. 
xs = np.arange(0,100,1)
ys = 200 + np.random.randn(100)

# matplotlib을 이용하여 그래프를 그립니다.
plt.plot(xs, ys)

# 그래프에 제목을 만들고 출력합니다. 
plt.title("Sample Visualization of Random Values")
plt.show()
```
