## 2.3 막대그래프: .plot(kind='bar')
- .value_counts(), .pivot_table()로 집계된 결과를 막대그래프로 표현할 수 있다. 
- .plot(kind='bar')  로 kind 인수에 그래프 종류를 지정할 수 있다. 
- 하나의 열(컬럼)은 하나의 범례(legend)를 구성한다. 
- 추가적으로 폰트 크기를 fontsize 로 지정할 수 있다. 
- 그래프의 크기를 figsize =(5,5) 와 같이 지정할 수 있다. 
- 막대그래프의 경우, stacked = True를 지정하면 적재된 형태로 표현된다. 

```python
# 2.3.1 봉사료 tips 데이터프레임에서 성별'sex' 데이터건수를 막대그래프로 시각화 해보자. 
# value_count & plot(kind='bar')

tips['sex'].value_counts()....;   # 집계된 결과에 그래프 작성 메소드를 연결 기술 한다. 
```

```python
# 2.3.2 봉사료 tips 데이터프레임에서 흡연여부별 'smoker' 데이터건수를 막대그래프로 시각화 해보자. 
# value_count & plot(kind='bar')

...;
```

```python
# 2.3.3  봉사료 tips 데이터프레임에서 성별 흡연여부별 'sex' x 'smoker' 데이터건수를 막대그래프로 시각화 해보자. 
# pivot_table & plot(kind='bar')
tips.pivot_table(values='tip',
                 index = 'sex',
                 columns= 'smoker',
                 aggfunc='count')....;
```

```python
# 2.3.4  봉사료 tips 데이터프레임에서 성별 흡연여부별 'sex' x 'smoker' 데이터건수를 막대그래프로 시각화 해보자. 
#        성별로 비교하되, 흡연여부를 구분하여 적재한다. (stacked=True)
#        퐅트 사이즈를 15로 지정한다. (fontsize = 15)
#        그래프 크기를 (5,5)인치로 지정한다. (figsize = (5,5))
# pivot_table & plot(kind='bar',...)
tips.pivot_table(values='tip',
                 index = 'sex',
                 columns= 'smoker',
                 aggfunc='count')....;
```

## 4.1 범주별 요약 통계값 집계
.pivot_table()의 집계함수를 이용해서 다양한 집계를 수행할 수 있다. 
- values = : 집계 할 수치변수(양적변수)
- index = : 행 방향으로 정리할 범주 (질적변수)
- columns = : 열 방향으로 정리할 범주 (질적변수)
- aggfunc = : 집계함수 ( 'count','mean','median','max','min' ,'std' 등)

인수를 지정하면서 그 개수에 따라 다양한 형태의 결과가 반환된다. 복수개가 지정 될 때에는 대괄호 안에 나열한다.[]
- values는 한 개 이상 복수개가 될 수 있다. 
- index와 columns는 합해서 1개 이상이 필요하며, 각각이 복수개가 될 수 있으며 결과는 계층 구조로 집계하여 반환 된다. 
- aggfunc는 지정하지 않으면 'mean'이 적용되고, [] 안에 복수개 지정할 수 도 있다.
